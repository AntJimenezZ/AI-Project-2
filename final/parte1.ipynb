```python
# ======================================================
#   MONTAR GOOGLE DRIVE
# ======================================================
from google.colab import drive
drive.mount('/content/drive')

import tensorflow as tf
from tensorflow.keras.applications import ResNet50, MobileNetV2
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import time
import pandas as pd


# ======================================================
#   NUEVA CONFIGURACIÓN DE RUTAS
# ======================================================

# Asegúrate de colocar tu dataset dentro de:
# /content/drive/MyDrive/tu_carpeta/data/train
# /content/drive/MyDrive/tu_carpeta/data/val

BASE_DIR = "/content/drive/MyDrive"
DATASET_DIR = f"{BASE_DIR}/data"
TRAIN_DIR = f"{DATASET_DIR}/train"
VAL_DIR   = f"{DATASET_DIR}/val"

SAVE_DIR = f"{BASE_DIR}/modelos"
os.makedirs(SAVE_DIR, exist_ok=True)



# ======================================================
#   CLASE PRINCIPAL
# ======================================================

class TransferLearningTrainer:
    def __init__(self, num_classes, img_size=224):
        self.num_classes = num_classes
        self.img_size = img_size
        self.model = None
        self.model_name = None
        self.train_generator = None
        self.val_generator = None
        self.class_names = None
        self.history = None

    def build_model(self, model_name='resnet50', trainable_layers=10):
        self.model_name = model_name

        if model_name == "resnet50":
            base_model = ResNet50(
                weights='imagenet', include_top=False,
                input_shape=(self.img_size, self.img_size, 3)
            )
        elif model_name == "mobilenetv2":
            base_model = MobileNetV2(
                weights='imagenet', include_top=False,
                input_shape=(self.img_size, self.img_size, 3)
            )
        else:
            raise ValueError("Modelo no válido")

        # Congelar capas
        for layer in base_model.layers[:-trainable_layers]:
            layer.trainable = False

        for layer in base_model.layers[-trainable_layers:]:
            layer.trainable = True

        x = GlobalAveragePooling2D()(base_model.output)
        x = Dense(512, activation='relu')(x)
        x = Dropout(0.5)(x)
        x = Dense(256, activation='relu')(x)
        x = Dropout(0.3)(x)
        predictions = Dense(self.num_classes, activation='softmax')(x)

        model = Model(inputs=base_model.input, outputs=predictions)

        model.compile(
            optimizer=tf.keras.optimizers.Adam(1e-4),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        self.model = model

    def prepare_data(self, train_dir, val_dir, batch_size=32):
        preprocessing = resnet_preprocess if self.model_name == 'resnet50' else mobilenet_preprocess

        train_gen = ImageDataGenerator(
            preprocessing_function=preprocessing,
            rotation_range=30,
            width_shift_range=0.25,
            height_shift_range=0.25,
            zoom_range=0.3,
            shear_range=0.25,
            brightness_range=[0.5, 1.5],
            horizontal_flip=True
        )

        val_gen = ImageDataGenerator(preprocessing_function=preprocessing)

        classes = sorted([d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))])

        self.train_generator = train_gen.flow_from_directory(
            train_dir,
            target_size=(self.img_size, self.img_size),
            batch_size=batch_size,
            class_mode='categorical',
            classes=classes
        )

        self.val_generator = val_gen.flow_from_directory(
            val_dir,
            target_size=(self.img_size, self.img_size),
            batch_size=batch_size,
            class_mode='categorical',
            shuffle=False,
            classes=classes
        )

        self.class_names = {v: k for k, v in self.train_generator.class_indices.items()}

    def train(self, epochs=5):
        model_path = f"{SAVE_DIR}/best_model_{self.model_name}.keras"

        checkpoint = ModelCheckpoint(
            model_path,
            save_best_only=True,
            monitor="val_loss",
            mode="min"
        )

        early_stop = EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True
        )

        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.3,
            patience=2
        )

        start = time.time()

        history = self.model.fit(
            self.train_generator,
            validation_data=self.val_generator,
            epochs=epochs,
            callbacks=[checkpoint, early_stop, reduce_lr]
        )

        end = time.time()
        self.training_time = end - start
        self.history = history

        print(f"\nModelo guardado en:\n{model_path}")

        return history

    def evaluate(self, model_name):
        preds = self.model.predict(self.val_generator)
        y_pred = np.argmax(preds, axis=1)
        y_true = self.val_generator.classes

        cm = confusion_matrix(y_true, y_pred)

        plt.figure(figsize=(7, 6))
        sns.heatmap(
            cm,
            annot=True,
            fmt="d",
            xticklabels=self.class_names.values(),
            yticklabels=self.class_names.values(),
            cmap="Blues"
        )
        plt.title(f"Matriz de Confusión - {model_name}")
        plt.show()

        print(f"\n===== Classification Report - {model_name} =====\n")
        print(classification_report(y_true, y_pred, target_names=self.class_names.values()))

        return {
            "y_true": y_true,
            "y_pred": y_pred,
            "cm": cm
        }



# ======================================================
#   EJECUCIÓN Y COMPARACIÓN
# ======================================================

def count_classes(path):
    return len([
        d for d in os.listdir(path)
        if os.path.isdir(os.path.join(path, d)) and not d.startswith('.')
    ])

num_classes = count_classes(TRAIN_DIR)
print(f"Clases detectadas: {num_classes}")

results = {}
histories = {}

for model_name in ["resnet50", "mobilenetv2"]:
    print(f"\n========== ENTRENANDO {model_name.upper()} ==========\n")

    trainer = TransferLearningTrainer(num_classes=num_classes)
    trainer.build_model(model_name)
    trainer.prepare_data(TRAIN_DIR, VAL_DIR)

    history = trainer.train(epochs=10)

    histories[model_name] = history
    results[model_name] = {
        "time": trainer.training_time,
        "eval": trainer.evaluate(model_name)
    }


# ======================================================
#   COMPARACIÓN FINAL
# ======================================================

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
for m, h in histories.items():
    plt.plot(h.history["val_accuracy"], label=f"{m} val_acc")
plt.title("Comparación Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
for m, h in histories.items():
    plt.plot(h.history["val_loss"], label=f"{m} val_loss")
plt.title("Comparación Loss")
plt.legend()

plt.show()

df_summary = pd.DataFrame({
    "Modelo": ["ResNet50", "MobileNetV2"],
    "Tiempo Entrenamiento (s)": [
        results["resnet50"]["time"],
        results["mobilenetv2"]["time"]
    ],
    "Último Val Accuracy": [
        histories["resnet50"].history["val_accuracy"][-1],
        histories["mobilenetv2"].history["val_accuracy"][-1]
    ],
    "Mín Val Loss": [
        min(histories["resnet50"].history["val_loss"]),
        min(histories["mobilenetv2"].history["val_loss"])
    ]
})

print("\n=========== RESUMEN COMPARATIVO ===========\n")
print(df_summary)
```
